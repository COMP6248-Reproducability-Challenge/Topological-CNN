```python
import json
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
import re
```

```python
file_old = open('Checkpoint/2022_04_29_17_17_13_Synthetic_results.json')
file = open('Checkpoint/2022_04_29_18_15_32_Synthetic_results.json')
results = json.load(file)
results_old = json.load(file_old)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
fig, ax = plt.subplots(2,2)
x = np.linspace(0,3000,300)
ax[0,1].title.set_text('Noisy MNIST Test-set')
ax[0,0].title.set_text('Noisy MNIST Training-set')

for key in results:
    if 'train' in key:
        ax[0,0].plot(x,results[key]['testing_accuracy'], label=key)
        # ax[0,0].legend()

for key in results:
    if 'train' in key:
        ax[1,0].plot(x,results[key]['training_loss'], label=key)
        # ax[1,0].legend()

for key in results:
    if 'test' in key:
        ax[0,1].plot(x,results[key]['testing_accuracy'], label=key)
        # ax[0,1].legend()
for key in results:
    if 'test' in key:
        ax[1,1].plot(x,results[key]['training_loss'], label=re.sub(r'_noisy_test','',key))
        # ax[1,1].legend()
        
lines, labels = fig.axes[-1].get_legend_handles_labels()
    
fig.legend(lines, labels, loc = 'center right')
fig.
        
```

```python
plt.rcParams["figure.figsize"] = (15,5)
plt.rcParams.update({'font.size': 12})

fig, ax = plt.subplots(2,2)
x = np.linspace(0,3000,300)
ax[0,1].title.set_text('Noisy MNIST Test-set')
ax[0,0].title.set_text('Noisy MNIST Training-set')
ax[0,0].set_ylabel('Testing Accuracy')
ax[1,0].set_ylabel('Training Loss')
for key in results:
    if 'train' in key:
        ax[0,0].plot(x,results_old[key]['testing_accuracy'], label=key)
        # ax[0,0].legend()

for key in results:
    if 'train' in key:
        ax[1,0].plot(x,results_old[key]['training_loss'], label=key)
        # ax[1,0].legend()

for key in results:
    if 'test' in key:
        ax[0,1].plot(x,results_old[key]['testing_accuracy'], label=key)
        # ax[0,1].legend()
for key in results:
    if 'test' in key:
        ax[1,1].plot(x,results_old[key]['training_loss'], label=re.sub(r'_noisy_test','',key))
        # ax[1,1].legend()
        
lines, labels = fig.axes[-1].get_legend_handles_labels()
    
fig.legend(lines, labels, loc = 'center right')
fig.text(0.5, 0.04, 'Batches', ha='center', va='center');
plt.savefig('synthetic_results.pdf')
```
